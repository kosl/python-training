```python
# Scipy fit and optimization
from scipy import optimize
import numpy as np
```

```python
# X range
x_data = np.linspace(-5, 5, num=50)

# Sinus data with noise
y_data = 2.9 * np.sin(1.5 * x_data) + np.random.normal(size=50)
```

```python
# Visualizing the data
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x_data, y_data, 'o')
```

```python
# If we know the nature of the data, i.e., the current data is a sine wave, we 
# can find parameters for the function with the use of least squares methods.

def test_function(x, a, b):
    return a * np.sin(b * x)

# test_function will be used by the least squares method to find the most
# suitable parameters to plot though the data.

# Let's use the scipy.optimize.curve_fit to fit a curve through the data

params, params_covariance = optimize.curve_fit(test_function, x_data, y_data,
                                               p0=[2., 2.]) # p0 are the initial parameters
print(f"Parameters:\na={params[0]}\nb={params[1]}")
```

```python
# Now let's plot the original data against the fitted function.

plt.plot(x_data, y_data, '-', label='Data')

# Let's create the fitted data
fitted_data = [test_function(x, params[0], params[1]) for x in x_data]
plt.plot(x_data, fitted_data, label='Fitted data')
plt.legend()
         
```
